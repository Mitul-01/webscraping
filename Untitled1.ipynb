```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
source = requests.get(url).text
```

```python
source
```

```python
soup = BeautifulSoup(source, "html.parser")
```

```python
covid_cases = soup.find("div", class_="maincounter-number").text.split("\n")[1]
```

```python
covid_cases
```

```python
deaths = soup.find("div", class_="maincounter-number").span.text
```

```python
deaths
```

```python

table = soup.find("table", attrs = {"id": "main_table_countries_today"} )
```

```python
table
```

```python
header = []
```

```python
header = table.thead.find_all("tr")
```

```python
header
```

```python
names = []
for th in header[0].find_all("th"):
    names.append(th.text)
    
```

```python
names
```

```python
body = table.tbody.find_all("tr")
```

```python
body
```

```python
data = []
for i in range(1,len(body)):
    row = []
    for j in body[i].find_all("td"):
        row.append(j.text.replace("\n", "").strip())
    data.append(row)
    
```

```python
data
```

```python
df = pd.DataFrame(data,columns = names)
```

```python
df
```
